```python
#Assignment(05) #Task No.1

#definig a function to get factorial of a number

#Defining the function
def fact(n):
    if n == 0: #setting the condition for the input number
        return 1
    else:
        return n*fact(n-1)

n = int(input("Enter a number here: "))
print("Factorial of", str(n), "is", end=" = ")

#calling the function
print(fact(n))

input("\n\n press enter to exit!")
```

```python
#Assignment(05) #Task No.2

#Defining a function to calculate numbers of uppercase and lowercase letters.


def uplow_test(x): #defining the function
    
    a = {"uppercases":0,"lowercases":0} #setting the no. of upper & lower cases to zero
    
    for y in x:
        
        if y.isupper(): #checking whether there is any uppercase letter
            
            a["uppercases"]+=1 #if yes, then add 1 into the value of uppercases in "a"
            
        elif y.islower(): #checking whether there is any lowercase letter
            
            a["lowercases"]+=1 #if yes, then add 1 into the value of lowercases in "a"
            
        else:
            
            pass
        
    print("Your input = ", x)
    print("There are", a["uppercases"], "Upper-case letters in your input")
    print("AND")
    print("There are", a["lowercases"], "Lower-case letters in your input")
        
x = input("Enter Your sentence here: ")

#calling the function
uplow_test(x)
        
    
#exiting
input("Press Enter to exit!")
```

```python
#Assignment(05) #TASK No.3

#Defining a funtion to get the even numbers from a given list
def even():
    list1 = [10, 21, 4, 45, 66, 93] 
    print("The Even numbers are the folowings")
 
    for num in list1: #iterating each number in list
      
    #checking condition whether the each num is divisible by "2"
        if num % 2 == 0: 
            print(num, end = "  ")

#Calling the function:     
even()

#exiting
input("\npress enter to exit!")
```

```python
#ASSIGNMENT(05)
#TASK no.4

#defining a function to check whether the user's string is palindrome or not.

def pal(x): #defining the function.
    string = "" 
    for i in x: 
        string = i + string 
    if (x==string): 
        print("YES\n The string is a palindrome!")
    else:
        print("\nNot a palindrome")

print("-------------------------")            
x = input("Enter a string here: ")#getting a string from user.
print("-------------------------")

#calling the function.
pal(x)

#exiting
input("press enter to exit!")
```

```python
#Assignment (05)
#Task No.05

def prime(): #Defining the function!
    
    number = int(input("Enter any number here: ")) #getting a number from the user.


    #setting a condition.
    
    if number > 1:
        for i in range(2, number):
            if (number % i) == 0:
                print(number, "is not a prime number")
                break
            



        else:
            print(number, "is a prime number")
#calling the function by its name            
prime()
```
